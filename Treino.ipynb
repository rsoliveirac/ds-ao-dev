## Modulo 4 - requisição da API

```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
url = 'https://www.magazineluiza.com.br/busca/smartwatch/'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
```

```python
page = requests.get(url, headers = headers)
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
products = soup.find('div', class_ = 'productShowCaseContent' )
```

```python
products_list = products.find_all( 'li',  class_ = 'product' )
```

```python
#id product
id_products = [i.get('id') for i in products_list]

```

```python
products_list = products.find_all ('span', class_ = 'price')
products_list[0].get_text('price-value')

price_products = [i.get_text('price-value') for i in products_list]
```

```python
products_list = products.find_all ('h3', itemprop = 'name')
products_list[0].get_text('name')

description_products = [i.get_text('name') for i in products_list]
```

```python
df = pd.DataFrame([id_products, description_products, price_products]).T

df.columns = ['ID', 'price', 'description']

df
```
