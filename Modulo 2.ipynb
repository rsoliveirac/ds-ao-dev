# Aula 1.0. Pensamento Analítico

## Etapas do pensamento analítico - ferramentas ou etapas para resolver um problema

1. Identificação da causa raíz 
2. Definir um escopo fechado para uma pergunta aberta (limites para criar a solução)
   - Limite de produto
   - limite de tempo
   - limite de preço
   - limite de localização

3. Quebrar o problema definido em tarefas menores
4. Organizar as tarefas por ordem lógica
5. Executar com uma mentalidade cíclica 



## Exemplo

Problema de negócio: Qual é o melhor preço de venda do produto?

1. Identificação da causa raíz 

   - Motivação: Qual o contexto?

     - A empresa está entrando no mercado de moda dos USA e não tem experiência para precificar o produto

   - Qual é a causa raíz do problema?

     - Preço de produto

     - Preço ótimo para maximizar o lucro

       

2. Definir um escopo fechado para uma pergunta aberta (limites para criar a solução)

   **Pergunta** **aberta**: Qual é o melhor preço de venda do produto?

   **Como transformar em pergunta fechada:** Mediana dos preços dos sites concorrentes [LIMITE DO PRODUTO] por produto, tipo e cor [LIMITE DO PRODUTO] dos últimos 30 dias [LIMITE DE TEMPO]. 

   <u>Dúvida: Isso é discutido com o cliente / time de negócio?</u>

   

3. Quebrar o problema definido em tarefas menores

   **Mediana dos preços dos sites concorrentes por produto, tipo e cor dos últimos 30 dias.** 

   - ​	Tarefas:
     - [ ] Calcular a mediana de preço dos sites concorrentes por produto, tipo e cor dos últimos 30 dias. 
     - [ ] Montar uma base de dados que contenha informações do produto, preço, tipo, cor, dias de exposição.
     - [ ] Definir o schema (colunas da tabela)
     - [ ] Define a infraestrutura (banco de dados, csv, API)
     - [ ] Design de ETL
     - [ ] Agendamento da atualização da tabela
     - [ ] Calcular a mediana e entrega do produto final

4. Organizar as tarefas por ordem lógica

   1. Montar uma base de dados que contenha informações do produto, preço, tipo, cor, dias de exposição
   2.  Definir o schema (colunas da tabela)
   3. Define a infraestrutura (banco de dados, csv, API)
   4. Design de ETL
   5. Calcular a mediana de preço dos sites concorrentes por produto, tipo e cor dos últimos 30 dias. 
   6. Fazer visualização do produto final
   7. Entrega do produto final 

5. Executar com uma mentalidade cíclica 

   - ​	Preciso passar por toda as tarefas o mais rapido possivel para:

     ​					- Identificar bloqueios 

     ​					- Indetificar impeditivos que possam desvalidar o projeto

     ​					- Entregar o valor para empresa rapidamente

   - Fazer escolhas simples e depois ir aperfeiçoando

# Aula 3.0. Método SAPE [saída - processo - entrada ]

## O problema de negócio

Dois sócios estão inovando e querem concorrer com as empresas americanas H&M e Macys. Ambos não tem experiência no mercado e por isso contrataram uma equipe de DS para responder a três perguntas:

1. Qual é o melhor preço de venda para as calças?
2. Quantos tipos de caças e suas cores para o produto inicial?
3. Quais as matérias-prima ncessárias para confeccionar calças?



## Planejamento da solução - O  método SAPE

**Qual o problema de negócio?**

 - Qual o melhor preço de venda para calças?

1. Pensar na saída (produto final):
   - A resposta para a pergunta - pensa sempre em temros mais simples como média, mediana e quartis
   - Formato da entrega - O que a pessoa quer ver? Gráficos, tabelas?
   - Local de entrega - e-mail? Telagram?
2. Processo (passo a passo):
   1. passo a paso para construir o cálculo da mediana ou média
   2. Definir o formato da entrega (Visualizações, tabelas, frases)
   3. Decidir o local de entrega (PowerBI, Telegram, email?)
3. Entrada (Fonte de dados):
   1. Fonte de dados
   2. Ferramentas

# Aula 4.0 Método SAPE na prática

## Planejamento da solução - O  método SAPE

1. **Qual o problema de negócio?**

 - Qual o melhor preço de venda para calças?

2. Pensar na saída (produto final):

- A resposta para a pergunta 

  - Mediana dos preços dos concorrentes 

- Formato da entrega 

  - Tabela ou Gráfico 

    - Criar simulação com o excel ou alguma outra de planilhas com valores aleatórios

      

- Local de entrega - e-mail? Telagram?

  - App Streamlit

3. **Processo (passo a passo):**
   1. passo a paso para construir o cálculo da mediana ou média
      -  Realizar o calculo da mediana sobre  o produto, tipo e cor
   2. Definir o formato da entrega (Visualizações, tabelas, frases)
      1.  Gráfico de barras com a mediana dos preços dos produtos por tipo e cor dos últimos 30 dias
      2. Tabela com as seguintes colunas:  id | nome_produto | tipo_produto | cor_produto | preco_produto
      3. Definição do Schema : colunas e seu tipo
      4. Definição e infraestrutura de armazenamento
      5. planejamento de agendamento dos scripts 
      6. fazer as visualizações
      7. entrega do produto final
   3. Decidir o local de entrega (PowerBI, Telegram, email?)
      - app com streamlit

4. **Entrada (Fonte de dados):**

- Fonte de dados

  ​	site da H&M e site da Macys

- Ferramentas

  python 3.8.0

  bibliotecas de webscrapping (bs4, selenium)

  pycharm

  Jupyter notebook (analises e prototipagens)

  crontjov, airflow
